## Calculate key statistics

### Subtask:
Calculate the average age, the percentage of individuals with different types of investments, and find the most common savings objectives.


**Reasoning**:
Calculate the average age, investment percentages, and savings objectives counts as requested by the subtask.



```python
average_age = df['age'].mean()

investment_columns = ['Mutual_Funds', 'Equity_Market', 'Debentures', 'Government_Bonds', 'Fixed_Deposits', 'PPF', 'Gold']
investment_percentages = {}
for col in investment_columns:
    percentage = (df[col] > 0).mean() * 100
    investment_percentages[col] = percentage

savings_objectives_counts = df['What are your savings objectives?'].value_counts()

display(average_age)
display(investment_percentages)
display(savings_objectives_counts)
```

## Visualize key statistics

### Subtask:
Create a visualization for the distribution of age.


**Reasoning**:
Create a histogram to visualize the distribution of age.



```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age', bins=10, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

## Visualize key statistics

### Subtask:
Create visualizations to show the percentages of different investment types.


**Reasoning**:
Create a bar plot to visualize the investment percentages dictionary, set the title and axis labels, and rotate x-axis labels for readability.



```python
plt.figure(figsize=(12, 7))
plt.bar(investment_percentages.keys(), investment_percentages.values(), color='skyblue')
plt.title('Percentage of Individuals with Different Investment Types')
plt.xlabel('Investment Type')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Visualize key statistics

### Subtask:
Create a visualization to show the counts of different savings objectives.


**Reasoning**:
Generate a bar plot to visualize the counts of different savings objectives using the pre-calculated `savings_objectives_counts` Series and set the plot's title and axis labels for clarity. Rotate the x-axis labels and adjust the layout for better readability.



```python
plt.figure(figsize=(10, 6))
sns.barplot(x=savings_objectives_counts.index, y=savings_objectives_counts.values)
plt.title('Counts of Different Savings Objectives')
plt.xlabel('Savings Objective')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Summarize findings

### Subtask:
Present the calculated statistics and insights from the visualizations.


**Reasoning**:
Summarize the calculated statistics and insights from the visualizations as instructed.



```python
print(f"Summary of Financial Data Analysis:")
print(f"\nAverage Age: {average_age:.2f} years")
print("\nInvestment Percentages:")
for investment, percentage in investment_percentages.items():
    print(f"- {investment}: {percentage:.2f}%")
print("\nSavings Objectives:")
for objective, count in savings_objectives_counts.items():
    print(f"- {objective}: {count}")
print("\nObservations from Visualizations:")
print("- The bar chart for investment percentages shows that all individuals in this dataset have some level of investment in all listed categories.")
print("- The bar chart for savings objectives indicates that 'Retirement Plan' is the most common objective, followed by 'Health Care' and 'Education'.")
print("- The histogram of age distribution shows the frequency of different ages in the dataset, with apparent peaks indicating the most common age ranges.")
```
